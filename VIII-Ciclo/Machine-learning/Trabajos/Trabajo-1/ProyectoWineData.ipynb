```python
def drop_outliers(df, col):
  q25, q75 = df[col].quantile(0.25), df[col].quantile(0.75)
  iqr = q75 - q25

  max = q75 + (1.5 * iqr)
  min = q25 - (1.5 * iqr)

  for i, c in enumerate(df[col]):
    if c < min or c > max:
      df.loc[i, col] = np.nan
```

```python
for column_name in datawhite:
  drop_outliers(datawhite, column_name)

for column_name in datared:
  drop_outliers(datared, column_name)
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn import metrics

def polinomial_regression(df, target, degree, alpha):
    train = df.sample(frac=0.8, random_state=123)
    test = df.drop(train.index).sample(frac=1, random_state=123)

    x_train = train.drop(columns=target)
    y_train = train[target]

    x_test = test.drop(columns=target)
    y_test = test[target]

    scaler = StandardScaler()
    x_train_scale = scaler.fit_transform(x_train)
    x_test_scale = scaler.transform(x_test)

    poly = PolynomialFeatures(degree=degree)
    x_train_poly = poly.fit_transform(x_train_scale)
    x_test_poly = poly.transform(x_test_scale)

    modelo = Lasso(alpha=alpha)
    # modelo = Ridge(alpha=alpha)
    # modelo = LinearRegression()
    modelo.fit(x_train_poly, y_train)

    y_pred = modelo.predict(x_test_poly)

    print("Train score: {}".format(modelo.score(x_train_poly, y_train)))
    print("Test score: {}".format(modelo.score(x_test_poly, y_test)))
    print()
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

    plt.figure()
    plt.plot(x_train_poly, y_train, 'bo')
    plt.plot(x_test_poly, y_pred, 'r-')
```

```python
polinomial_regression(df=datawhite, target='quality', degree=1, alpha=1.0)
```

```python
polinomial_regression(df=datared, target='quality', degree=1, alpha=0.0)
```
