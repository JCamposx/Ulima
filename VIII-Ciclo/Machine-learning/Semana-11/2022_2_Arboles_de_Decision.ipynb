```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
data.dropna(inplace=True)

X = data.drop(columns='species')
Y = data['species']
X = pd.get_dummies(X, columns=['island', 'sex'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier

arbol = DecisionTreeClassifier(min_samples_leaf=50)
arbol.fit(x_train, y_train)

print(f"Accuracy")
print(f"Training: {arbol.score(x_train, y_train):.4f}")
print(f"Testing: {arbol.score(x_test, y_test):.4f}")
```

```python
arbol.get_params()
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(50, 20))
plot_tree(arbol, feature_names=X.columns, class_names=Y.unique())
plt.show()
```

```python
data = pd.read_csv('./CafeData.csv')
X = data.drop(columns='LEER')
Y = data['LEER']
arbol_cafe = DecisionTreeClassifier()
arbol_cafe.fit(X, Y)
plt.figure(figsize=(50, 20))
plot_tree(arbol_cafe, feature_names=X.columns)
plt.show()
```
