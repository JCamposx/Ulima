```python
import numpy as np
from matplotlib import animation as anim
from matplotlib import rc
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
rc('animation', html='jshtml')
```

```python
x, y = make_classification(n_samples=20, n_classes=2, n_clusters_per_class=1,
                           n_features=1, n_informative=1, n_redundant=0, n_repeated=0, class_sep=1)

print(x.shape)

plt.figure()
plt.plot(x, y, "b.")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
fig = plt.figure(figsize=(6, 4))
ax = plt.axes(xlim=(np.amin(x), np.amax(x)), ylim=(-0.1, 1.1))
line1, = ax.plot([], [], 'g', lw=2)
line0, = ax.plot([], [], 'b', lw=2)
x0 = np.linspace(np.amin(x), np.amax(x), 1000)
x0 = x0.reshape(x0.shape[0], 1)

modelo = SGDClassifier(loss='log', alpha=0, learning_rate='constant', eta0=0.1)


def init():
    line1.set_data([], [])
    line0.set_data([], [])
    ax.plot(x, y, "b.")
    return line1, line0,


def animate(i):
    modelo.partial_fit(x_train, y_train, classes=np.unique(y_train))
    line1.set_data(x0, modelo.predict_proba(x0)[:, 1])
    line0.set_data(x0, modelo.predict_proba(x0)[:, 0])
    return line1, line0,
```

```python
anim.FuncAnimation(fig, animate, init_func=init,
                   frames=100, interval=20, blit=True)
```
